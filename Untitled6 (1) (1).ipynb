```python
import pandas as pd
import numpy as np
from fuzzywuzzy import process
import warnings
warnings.filterwarnings("ignore")
```

```python
movies=pd.read_csv("movies.csv")
```

```python
ratings=pd.read_csv("ratings.csv")
```

```python
ratings.head()
```

```python
arjun=pd.merge(movies,ratings,on="movieId")
```

```python
arjun.head()
```

```python
arjun.describe()
```

```python
arjun=arjun.drop(columns=["timestamp"])
```

```python
arjun.head()
```

```python
arjuna=pd.DataFrame(arjun.groupby("title")["rating"].mean())
```

```python
arjuna.head()
```

```python
arjuna["number of ratings"]=arjun.groupby("title")["rating"].count()
```

```python
arjuna.head()                     
```

```python
moviematrix=arjun.pivot_table(index="title",columns="userId",values="rating").fillna(0)
```

```python
moviematrix.head()
```

```python

from scipy.sparse import csr_matrix

moviematrixx = csr_matrix(moviematrix.values)

from sklearn.neighbors import NearestNeighbors
knn=NearestNeighbors(metric="cosine",algorithm='brute', n_neighbors=20)
knn.fit(moviematrixx)
```

```python
moviematrix.shape
```

```python

def recommender(movie_name, data,model, n_recommendations ):
    model.fit(data)
    inox=process.extractOne(movie_name, movies['title'])[2]
    print('Movie Selected: ',movies['title'][inox], 'Index:',inox)
    print('Searching for recommendations.....')
    distances, indices=model.kneighbors(data[inox],n_neighbors=25)
    for i in indices:
        print(movies['title'][i].where(i!=inox))
    
recommender('hellboy',moviematrixx,knn,25)
```
